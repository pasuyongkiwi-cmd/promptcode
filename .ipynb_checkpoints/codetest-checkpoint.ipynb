```python
import os
from flask import Flask, render_template, request
from dotenv import load_dotenv

load_dotenv()

# ===== 先尝试导入外部模块；如果还没创建，就用临时占位函数 =====
try:
    from rules import (
        check_syntax,
        run_flake8,
        check_docstrings,
        check_naming,
        build_rule_summary,
    )
except ModuleNotFoundError:
    def check_syntax(code):
        try:
            compile(code, "<string>", "exec")
            return {"passed": True, "error": ""}
        except SyntaxError as exc:
            return {
                "passed": False,
                "error": f"SyntaxError: {exc.msg} (line {exc.lineno})"
            }

    def run_flake8(code):
        return {
            "passed": True,
            "issues": [],
            "error": "rules.py not created yet, flake8 skipped.",
            "available": False,
        }

    def check_docstrings(code):
        return {
            "passed": True,
            "missing_docstrings": [],
            "error": "rules.py not created yet, docstring check skipped.",
        }

    def check_naming(code):
        return {
            "passed": True,
            "issues": [],
            "error": "rules.py not created yet, naming check skipped.",
        }

    def build_rule_summary(syntax_result, docstring_result, naming_result, flake8_result):
        return {
            "syntax_check": syntax_result,
            "docstring_check": docstring_result,
            "naming_check": naming_result,
            "flake8_check": flake8_result,
        }

try:
    from reviewer import review_code_with_llm
except ModuleNotFoundError:
    def review_code_with_llm(code, mode, rule_summary):
        return {
            "summary": "reviewer.py not created yet, so this is a placeholder summary.",
            "reasoning": "LLM review is currently unavailable because reviewer.py has not been created.",
            "bugs": [],
            "style_issues": [],
            "maintainability_issues": [],
            "documentation_issues": [],
            "suggestions": ["Create reviewer.py next."],
            "improved_code": code,
            "generated_doc": "",
            "scores": {
                "correctness": 3,
                "readability": 3,
                "maintainability": 3,
                "documentation_quality": 3,
            },
            "overall_score": 3,
        }

try:
    from utils import normalize_mode, empty_llm_result, merge_review_results
except ModuleNotFoundError:
    def normalize_mode(mode):
        if not mode:
            return "full"
        mode = mode.strip().lower()
        if mode in {"bug", "bug_review"}:
            return "bug"
        if mode in {"style_docs", "style", "style_review", "style_docs_review"}:
            return "style_docs"
        return "full"

    def empty_llm_result():
        return {
            "summary": "",
            "reasoning": "",
            "bugs": [],
            "style_issues": [],
            "maintainability_issues": [],
            "documentation_issues": [],
            "suggestions": [],
            "improved_code": "",
            "generated_doc": "",
            "scores": {
                "correctness": 1,
                "readability": 1,
                "maintainability": 1,
                "documentation_quality": 1,
            },
            "overall_score": 1,
        }

    def merge_review_results(llm_result, rule_summary):
        return llm_result

app = Flask(__name__)
app.config["SECRET_KEY"] = os.getenv("SECRET_KEY", "dev-secret-key")


@app.route("/", methods=["GET"])
def index():
    return render_template("index.html")


@app.route("/about", methods=["GET"])
def about():
    return "<h1>About PromptCode</h1><p>LLM-powered Python code review tool.</p>"


@app.route("/review", methods=["POST"])
def review():
    code = request.form.get("code", "").strip()
    mode = normalize_mode(request.form.get("mode", "full"))

    if not code:
        return render_template(
            "result.html",
            code=code,
            mode=mode,
            error="Code input cannot be empty.",
            syntax_result={},
            docstring_result={},
            naming_result={},
            flake8_result={},
            rule_summary={},
            llm_result=empty_llm_result(),
            final_result=empty_llm_result(),
        )

    syntax_result = check_syntax(code)
    docstring_result = check_docstrings(code)
    naming_result = check_naming(code)
    flake8_result = run_flake8(code)

    rule_summary = build_rule_summary(
        syntax_result=syntax_result,
        docstring_result=docstring_result,
        naming_result=naming_result,
        flake8_result=flake8_result,
    )

    llm_result = review_code_with_llm(
        code=code,
        mode=mode,
        rule_summary=rule_summary,
    )

    final_result = merge_review_results(llm_result, rule_summary)

    return render_template(
        "result.html",
        code=code,
        mode=mode,
        error=None,
        syntax_result=syntax_result,
        docstring_result=docstring_result,
        naming_result=naming_result,
        flake8_result=flake8_result,
        rule_summary=rule_summary,
        llm_result=llm_result,
        final_result=final_result,
    )


if __name__ == "__main__":
    app.run(debug=False, use_reloader=False, port=5000)
```
